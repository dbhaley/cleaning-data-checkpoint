```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df= pd.read_csv('hospital_readmissions.csv')
df
```

```python
#Although i did not find any inconsistencies, I Changed long names to condense the data/improve visual evaluation.
df = df.rename(columns={'time_in_hospital':'days','n_outpatient':'opv','n_inpatient':'inptv','diabetes_med':'diab_med','n_lab_procedures':'nlabs','n_procedures':'nproc','n_medications':'meds','n_emergency':'erv','medical_specialty':'mdtype','glucose_test':'g_test','readmitted':'readmit'})
df
```

```python
#Would like to analyze categorical data, together with numerical data, to see the big picture.  Will convert categorical data to numerical data
#First will convert g_test

df['g_test'].replace(['no','normal','high'],[0,1,2], inplace = True)
df
```

```python
#Would like to analyze categorical data, together with numerical data, to see the big picture.  Will convert categorical data to numerical data
# Will convert A1C_test

df['A1Ctest'].replace(['no','normal','high'],[0,1,2], inplace = True)
df
```

```python
#replace the values for "change" (of meds)
df['change'].replace(['no','yes'],[0,1], inplace = True)
df
```

```python
#replace the values diabetes_meds to binary as well...
df['diab_med'].replace(['no','yes'],[0,1], inplace = True)
df
```

```python
#replace the diagnosis with numerical values
df['diag_1'].replace(['Circulatory','Other','Injury','Digestive', 'Respiratory','Diabetes', 'Musculoskeletal', 'Missing'],[1,2,3,4,5,6,7,8], inplace = True)
df
```

```python
#replace the diagnosis with numerical values
df['diag_2'].replace(['Circulatory','Other','Injury','Digestive', 'Respiratory','Diabetes', 'Musculoskeletal', 'Missing'],[1,2,3,4,5,6,7,8], inplace = True)
df
```

```python
#replace the diagnosis with numerical values
df['diag_3'].replace(['Circulatory','Other','Injury','Digestive', 'Respiratory','Diabetes', 'Musculoskeletal', 'Missing'],[1,2,3,4,5,6,7,8], inplace = True)
df
```

```python
#replace medical specialty with numerical values
df['mdtype'].replace(['Missing','Other','InternalMedicine','Family/GeneralPractice','Cardiology','Surgery', 'Emergency/Trauma'],[1,2,3,4,5,6,7], inplace = True)
df
```
